NUMPY ARRAYS

* Unlike the lists, we cannot have different data types in the array.
* If we try to do something like that, numpy array will automatically up cost all the reamining integers to float. 

```python
np.array([1,4,2,5,3])
```

```python
np.array([1.4,4,2,3])
```

You can explicitly set the data type of array

```python
np.array([1,2,3,4,5,6],dtype = 'float32')
```
